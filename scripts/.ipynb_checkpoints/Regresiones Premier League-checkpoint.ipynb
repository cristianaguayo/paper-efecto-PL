```python
def significancia_variables(mod,var,replace_spaces = True, std_err = True, stars = True):
    coef = "%.3f" % (np.round(mod.params[var], decimals = 3))
    se = "%.3f" % (np.round(mod.bse[var], decimals = 3))
    if mod.params[var] > 0:
        output = '~' + str(coef)                 
    else:
        output = str(coef)
    if std_err:
        output = output + ' (' + str(se) + ')'
    if stars:
        pval = np.round(mod.pvalues[var], 3)
        if pval < 0.01:
            output = output + '***'
        elif pval < 0.05 and pval >= 0.01:
            output = output + '**~'
        elif pval < 0.1 and pval >=0.05:
            output = output + '*~~'
        else:
            output = output + '~~~'
    if replace_spaces:
        output = output.replace("~"," ")
    return output
```

```python
def ResumenRegresion1Var(df_reg,
                         variable,
                         r,
                         y = 'position',
                         ratio_lv = True):
    if ratio_lv:
        ratio = 'ratio_lv_p%s' % r
        if '%s' in y:
            formulas = [y % r + ' ~ %s + ' + ratio,
                        y % r + ' ~ %s + equipo + ' + ratio,
                        y % r + ' ~ %s + Torneo + ' + ratio,
                        y % r + ' ~ %s + equipo + Torneo + ' + ratio]
        else:
            formulas = [y + ' ~ %s + ' + ratio,
                        y + ' ~ %s + equipo + ' + ratio,
                        y + ' ~ %s + Torneo + ' + ratio,
                        y + ' ~ %s + equipo + Torneo + ' + ratio]
        regressors = ['Intercept',ratio, variable]
    else:
        if '%s' in y:
            formulas = [y % r + ' ~ %s',
                        y % r + ' ~ %s + equipo',
                        y % r + ' ~ %s + Torneo',
                        y % r + ' ~ %s + equipo + Torneo']
        else:
            formulas = [y + ' ~ %s',
                        y + ' ~ %s + equipo',
                        y + ' ~ %s + Torneo',
                        y + ' ~ %s + equipo + Torneo']
        regressors = ['Intercept', variable]   
    models = []
    for formula in formulas:
        formula_ = formula % variable
        mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                         cov_kwds={'groups': df_reg['equipo']})
        models.append(mod)
    tble = summary_col(models,
                       regressor_order = regressors,
                       drop_omitted=True,
                       stars = True,
                       float_format = '%0.3f',
                       model_names = ['Base','EF Equipo','EF Torneo', 'EFs Equipo y Torneo'],
                       info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                                  'R2':lambda x: "{:.2f}".format(x.rsquared),
                                  'R2a': lambda x: "{:.2f}".format(x.rsquared_adj),
                                  'LL': lambda x: "{:.2f}".format(x.llf)})
#     tble.tables[0] = tble.tables[0].loc[regressors]
    str_latex = tble.tables[0].to_latex()
    return str_latex
```

```python
def ConstructorTablaCoeficientes1VarLR(df_reg,
                                       variable,
                                       y = 'position',
                                       ratio_lv = True):
    rl = []
    columnas = ['Base','EF Equipo', 'EF Torneo', 'EF Equipo y Torneo']
    df_coef = pd.DataFrame(data = np.zeros((12,4)), columns = columnas)
    coefs = [[] for i in range(4)]
    for r in [4,5,6,7]:
        if ratio_lv:
            ratio = 'ratio_lv_p%s' % r
            if '%s' in y:
                formulas = [y % r + ' ~ %s + ' + ratio,
                            y % r + ' ~ %s + equipo + ' + ratio,
                            y % r + ' ~ %s + Torneo + ' + ratio,
                            y % r + ' ~ %s + equipo + Torneo + ' + ratio]
            else:
                formulas = [y + ' ~ %s + ' + ratio,
                            y + ' ~ %s + equipo + ' + ratio,
                            y + ' ~ %s + Torneo + ' + ratio,
                            y + ' ~ %s + equipo + Torneo + ' + ratio]
            regressors = ['Intercept',ratio, variable]
        else:
            if '%s' in y:
                formulas = [y % r + ' ~ %s',
                            y % r + ' ~ %s + equipo',
                            y % r + ' ~ %s + Torneo',
                            y % r + ' ~ %s + equipo + Torneo']
            else:
                formulas = [y + ' ~ %s',
                            y + ' ~ %s + equipo',
                            y + ' ~ %s + Torneo',
                            y + ' ~ %s + equipo + Torneo']
        for l in [3,4,5]:
            var = variable % (r,l)
            rl.append("$r = %s, l = %s$" % (r,l))
            for i in range(len(formulas)):
                formula_ = formulas[i] % var
                mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                                 cov_kwds={'groups': df_reg['equipo']})
                coefs[i].append(significancia_variables(mod,var, replace_spaces=False, std_err=True, stars = False))
    for i in range(4):
        df_coef[columnas[i]] = coefs[i]
    df_coef['rl'] = rl
    df_coef.set_index('rl', inplace = True)
    return df_coef
```

```python
def ConstructorCoeficientes1VarLR(df_reg,
                                  variable,
                                  formula,
                                  conf = 0.05,
                                  ratio_lv = True):
    rl = []
    coefs = []
    lb = []
    ub = []
    for r in [4,5,6,7]:
        for l in [3,4,5]:
            var = variable % (r,l)
            rl.append("$r = %s, l = %s$" % (r,l))
            if ratio_lv:
                ratio = 'ratio_lv_p%s' % r
                if formula.count('%s') > 1:
                    strreplace = (r,var)
                else:
                    strreplace = var
                formula_ = formula % strreplace + ' + ' + ratio
            else:
                if formula.count('%s') > 1:
                    strreplace = (r,var)
                else:
                    strreplace = var
                formula_ = formula % strreplace
            mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                             cov_kwds={'groups': df_reg['equipo']})
            coefs.append(mod.params[var])
            ci = mod.conf_int(alpha=conf)
            lb.append(ci.loc[var,0])
            ub.append(ci.loc[var,1])
    df_coef = pd.DataFrame(data = {'LB':lb, 'Beta' : coefs, 'UB':ub}, index = rl)
    return df_coef
```

```python
def GraficoCoeficientes1VarLR(datadir,
                              archivo,
                              formulas,
                              textformulas,
                              variable,
                              plotname,
                              conf = 0.05,
                              ratio_lv = True):
    dictreplaces = {'perc': 'Frac. partidos', 'facil': 'fáciles', 'dificil': 'dificiles','_':' ', 'general': ''}
    cint = str(int((1-conf)*100)) + '%'
    title = variable.split('_p')[0]
    for word, replace in dictreplaces.items():
        title = title.replace(word, replace)
    title = '$x_{it}$: ' + title
    df_reg = pd.read_excel(os.path.join(datadir, archivo))
    fig, axs = plt.subplots(figsize=(5,10), nrows = len(formulas), sharex= True)
    plt.suptitle(title)
    for i in range(len(formulas)):
        formula = formulas[i]
        textformula = textformulas[i]
        ax = axs[i]
        df_coef = ConstructorCoeficientes1VarLR(df_reg, variable, formula, conf, ratio_lv)
        x = df_coef.index.tolist()
        lb = df_coef.LB.tolist()
        ub = df_coef.UB.tolist()
        beta = df_coef.Beta.tolist()
        ax.plot(x, beta, color = 'g' ,label='Beta')
        ax.fill_between(x, lb, ub, color='grey', label='I.C. %s' %cint)
#         ax.plot(df_coef.index.tolist(), ,linestyle = ':', color = 'r' ,label='LB %s' % cint)
#         ax.plot(df_coef.index.tolist(), df_coef.UB.tolist(),linestyle = ':', color = 'b' ,label='UB %s' % cint)
        ax.axhline(y=0, color='k')
        ax.set_ylabel(r'$\beta$')
        ax.legend()
        ax.set_title(textformula)
        if i == len(formulas) - 1:
            ax.set_xlabel('Valores de $r$ y $l$')
    fig.tight_layout()
    fig.subplots_adjust(top=0.92)
    plt.xticks(rotation=90)
    fig.savefig(os.path.join(os.path.pardir,'Graficos','%s.png' % plotname), bbox_inches = "tight", dpi = 250)
```

```python
def ResumenRegresion1VarHet(df_reg,
                            variable,
                            r,
                            categoria,
                            y = 'position',
                            ratio_lv = True):
    if ratio_lv:
        ratio = 'ratio_lv_p%s' % r
        formulas = [y + ' ~ %s + ' + ratio,
                    y + ' ~ %s + equipo + ' + ratio,
                    y + ' ~ %s + Torneo + ' + ratio,
                    y + ' ~ %s + equipo + Torneo + ' + ratio]
        regressors = ['Intercept',
                      variable,
                      ratio,
                      categoria + '[T.facil]',
                      categoria + '[T.dificil]', 
                      variable + ':' + categoria + '[T.facil]',
                      variable + ':' + categoria + '[T.dificil]']
    else:
        formulas = [y + ' ~ %s',
                    y + ' ~ %s + equipo',
                    y + ' ~ %s + Torneo',
                    y + ' ~ %s + equipo + Torneo']
        regressors = ['Intercept',
                      variable,
                      categoria + '[T.facil]',
                      categoria + '[T.dificil]', 
                      variable + ':' + categoria + '[T.facil]',
                      variable + ':' + categoria + '[T.dificil]']
    df_reg[categoria] = df_reg[categoria].astype('category')
    df_reg[categoria] = df_reg[categoria].cat.reorder_categories(['regular', 'facil','dificil'])
    models = []
    inter = '%s*%s' % (variable,categoria)
    for formula in formulas:
        formula_ = formula % inter
        mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                         cov_kwds={'groups': df_reg['equipo']})
        models.append(mod)
    tble = summary_col(models,
                       regressor_order = regressors,
                       drop_omitted=True,
                       stars = True,
                       float_format = '%0.3f',
                       model_names = ['Base','EF Equipo','EF Torneo', 'EFs Equipo y Torneo'],
                       info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                                  'R2':lambda x: "{:.2f}".format(x.rsquared),
                                  'R2a': lambda x: "{:.2f}".format(x.rsquared_adj),
                                  'LL': lambda x: "{:.2f}".format(x.llf)})
    str_latex = tble.tables[0].to_latex()
    return str_latex
```

```python
def ConstructorTablaCoeficientes1VarHetLR(df_reg,
                                          variable,
                                          interaccion,
                                          y = 'position',
                                          ratio_lv = True):
    rl = []
    columnas = ['Base','EF Equipo', 'EF Torneo', 'EF Equipo y Torneo']
    coefs = [[] for i in range(4)]

    for r in [4,5,6,7]:
        if ratio_lv:
            ratio = 'ratio_lv_p%s' % r
            if '%s' in y:
                formulas = [y % r + ' ~ %s + ' + ratio,
                            y % r + ' ~ %s + equipo + ' + ratio,
                            y % r + ' ~ %s + Torneo + ' + ratio,
                            y % r + ' ~ %s + equipo + Torneo + ' + ratio]
            else:
                formulas = [y + ' ~ %s + ' + ratio,
                            y + ' ~ %s + equipo + ' + ratio,
                            y + ' ~ %s + Torneo + ' + ratio,
                            y + ' ~ %s + equipo + Torneo + ' + ratio]
        else:
            if '%s' in y:
                formulas = [y % r + ' ~ %s',
                            y % r + ' ~ %s + equipo',
                            y % r + ' ~ %s + Torneo',
                            y % r + ' ~ %s + equipo + Torneo']
            else:
                formulas = [y + ' ~ %s',
                            y + ' ~ %s + equipo',
                            y + ' ~ %s + Torneo',
                            y + ' ~ %s + equipo + Torneo']
        for l in [3,4,5]:
            var = variable % (r,l)
            cat = 'categoria_general_c%s' % l
            inter = var + '*' + cat
            df_reg[cat] = df_reg[cat].astype('category')
            df_reg[cat] = df_reg[cat].cat.reorder_categories(['regular', 'facil','dificil'])
            rl.append("$r = %s, l = %s$" % (r,l))
            dictentradas = {'xF' : var + ':' + cat + '[T.facil]',
                            'xD': var + ':' + cat + '[T.dificil]'}
            for i in range(len(formulas)):
                formula_ = formulas[i] % inter
                mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                                 cov_kwds={'groups': df_reg['equipo']})
                coefs[i].append(significancia_variables(mod,
                                                        dictentradas[interaccion],
                                                        replace_spaces=False,
                                                        std_err=True,
                                                        stars = False))
    df_coef = pd.DataFrame(data = np.zeros((12,4)), columns = columnas, index = rl)
    for i in range(len(columnas)):
        df_coef[columnas[i]] = coefs[i]
    return df_coef
```

```python
def ConstructorCoeficientes1VarHetLR(df_reg,
                                     variable,
                                     formula,
                                     conf = 0.05,
                                     ratio_lv = True):
    rl = []
    # Lista de listas: 5 listas con 3 entradas cada una.
    # Orden de la primera entrada: x, F, D, x*F, x*D
    # Orden de la segunda entrada: beta, lb, ub
    results = [[[], [], []] for i in range(5)]
    for r in [4,5,6,7]:
        for l in [3,4,5]:
            cat = 'categoria_general_c%s' % l
            var = variable % (r,l)
            inter = var + '*' + cat
            if ratio_lv:
                ratio = 'ratio_lv_p%s' % r
                if formula.count('%s') > 1:
                    strreplace = (r,inter)
                else:
                    strreplace = inter
                formula_ = formula % strreplace + ' + ' + ratio
            else:
                if formula.count('%s') > 1:
                    strreplace = (r,inter)
                else:
                    strreplace = inter
                formula_ = formula % strreplace
            df_reg[cat] = df_reg[cat].astype('category')
            df_reg[cat] = df_reg[cat].cat.reorder_categories(['regular', 'facil','dificil'])
            rl.append("$r = %s, l = %s$" % (r,l))
            entradas = [var,
                        cat + '[T.facil]',
                        cat + '[T.dificil]',
                        var + ':' + cat + '[T.facil]',
                        var + ':' + cat + '[T.dificil]']
            mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                             cov_kwds={'groups': df_reg['equipo']})
            ci = mod.conf_int(alpha=conf)
            for i in range(5):
                ent = entradas[i]
                results[i][0].append(mod.params[ent])
                results[i][1].append(ci.loc[ent,0])
                results[i][2].append(ci.loc[ent,1])
    dfs = []
    variables = ['x','F','D','xF','xD']
    for i in range(5):
        col = variables[i]
        data = results[i]
        df = pd.DataFrame({col: data[0], col + '_lb': data[1], col + '_ub': data[2]}, index = rl)
        dfs.append(df)
    df_coef = pd.concat(dfs, axis = 1)
    return df_coef
```

```python
def GraficoCoeficientes1VarHetLR(datadir,
                                 archivo,
                                 formulas,
                                 textformulas,
                                 variable,
                                 inters,
                                 plotname,
                                 conf = 0.05,
                                 ratio_lv = True):
    dictreplaces = {'perc': 'Frac. partidos', 'facil': 'fáciles', 'dificil': 'dificiles','_':' ', 'general': ''}
    dictvariables = {'x': '$x_{it}$',
                     'F' : '$F_{it}$',
                     'D': '$D_{it}$',
                     'xF':'$x_{it}F_{it}$',
                     'xD':'$x_{it}D_{it}$'}
    cint = str(int((1-conf)*100)) + '%'
    title = variable.split('_p')[0]
    for word, replace in dictreplaces.items():
        title = title.replace(word, replace)
    df_reg = pd.read_excel(os.path.join(datadir, archivo))
    dfs_coef = []
    for i in range(len(formulas)):
        formula = formulas[i]

        dfs_coef.append(ConstructorCoeficientes1VarHetLR(df_reg, variable, formula, conf, ratio_lv))


    for inter in inters:
        fig, axs = plt.subplots(figsize=(5,10), nrows = len(formulas), sharex= True)
        plt.suptitle(title + ' - ' + dictvariables[inter])
        for i in range(len(formulas)):
            df_coef = dfs_coef[i]
            formula = formulas[i]
            textformula = textformulas[i]
            ax = axs[i]
            x = df_coef.index.tolist()
            lb = df_coef[inter + '_lb'].tolist()
            ub = df_coef[inter + '_ub'].tolist()
            beta = df_coef[inter].tolist()
            ax.plot(x, beta, color = 'g' ,label='Beta')
            ax.fill_between(x, lb, ub, color='grey', label='I.C. %s' %cint)
    #         ax.plot(df_coef.index.tolist(), ,linestyle = ':', color = 'r' ,label='LB %s' % cint)
#         ax.plot(df_coef.index.tolist(), df_coef.UB.tolist(),linestyle = ':', color = 'b' ,label='UB %s' % cint)
            ax.axhline(y=0, color='k')
            ax.set_ylabel(r'$\beta$')
            ax.legend()
            ax.set_title(textformula)
            if i == len(formulas) - 1:
                ax.set_xlabel('Valores de $r$ y $l$')
#             fig.tight_layout()
            fig.subplots_adjust(top=0.92)
            plt.xticks(rotation=90)
            plotname_ = plotname + inter
            fig.savefig(os.path.join(os.path.pardir,'Graficos','%s.png' % plotname_), bbox_inches = "tight", dpi = 250)
```

```python
#EJECUCION
datadir = os.path.join(os.pardir,
                       'Datos',
                       'Bases construidas',
                       'Bases main',
                       'Bases efecto')
archivo = 'Inglaterra.xlsx'
df_reg_test = pd.read_excel(os.path.join(datadir, archivo))

# Resultado principal
print(ResumenRegresion1VarHet(df_reg_test,
                              'perc_facil_general_p6_c4',
                              r = 6,
                              categoria = 'categoria_general_c4',
                              y = 'Var_p%s' % 6,
                              ratio_lv = False))
print(ResumenRegresion1VarHet(df_reg_test,
                              'perc_dificil_general_p6_c4',
                              r = 6,
                              categoria = 'categoria_general_c4',
                              y = 'Var_p%s' % 6,
                              ratio_lv = False))

# Anexos
variables = ['perc_facil_general_p%s_c%s',
             'perc_dificil_general_p%s_c%s']
replaces = {'\\textasciitilde ':'~', '\\$':'$'}
inters = ['xF', 'xD']
for variable in variables:
    for inter in inters:
        table_anexo = ConstructorTablaCoeficientes1VarHetLR(df_reg_test,
                                                            variable,
                                                            interaccion=inter,
                                                            y = 'Var_p%s',
                                                            ratio_lv = False).to_latex()
        for key,value in replaces.items():
            table_anexo = table_anexo.replace(key,value)
        print(table_anexo)
# Graficos
formulas = ['Var_p%s ~ %s',
            'Var_p%s ~ equipo + %s',
            'Var_p%s ~ Torneo + %s',
            'Var_p%s ~ equipo + Torneo + %s']
textformulas = ['Base','EF Equipo', 'EF Torneo', 'EF Equipo y Torneo']
inters = ['xF', 'xD']
plotnames = ['fracfaciles1varhetatting','fracdificiles1varhetatting']
for i in range(len(variables)):
    variable = variables[i]
    plotname = plotnames[i]
    GraficoCoeficientes1VarHetLR(datadir,
                                 archivo,
                                 formulas,
                                 textformulas,
                                 variable,
                                 inters,
                                 plotname,
                                 conf = 0.05,
                                 ratio_lv = False)
```

```python
def ResumenRegresion2Var(df_reg, variable1, variable2, y = 'position'):
    formulas = [y + ' ~ %s',
                y + ' ~ %s + equipo',
                y + ' ~ %s + Torneo',
                y + ' ~ %s + equipo + Torneo']
    models = []
    variables = '%s + %s' % (variable1, variable2)
    for formula in formulas:
        formula_ = formula % variables
        mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                         cov_kwds={'groups': df_reg['equipo']})
        models.append(mod)
    regressors = ['Intercept', variable1, variable2]
    tble = summary_col(models,
                       regressor_order = regressors,
                       drop_omitted=True,
                       stars = True,
                       float_format = '%0.3f',
                       model_names = ['Base','EF Equipo','EF Torneo', 'EFs Equipo y Torneo'],
                       info_dict={'$N$':lambda x: "{0:d}".format(int(x.nobs)),
                                  '$R^2$':lambda x: "{:.2f}".format(x.rsquared),
                                  '$R^2_a$': lambda x: "{:.2f}".format(x.rsquared_adj),
                                  '$LL$': lambda x: "{:.2f}".format(x.llf)})
#     tble.tables[0] = tble.tables[0].loc[regressors]
    return tble.tables[0]
```

```python
def ResumenRegresion2VarHet(df_reg, variable1, variable2, categoria, y = 'position'):
    formulas = [y + ' ~ %s',
                y + ' ~ %s + equipo',
                y + ' ~ %s + Torneo',
                y + ' ~ %s + equipo + Torneo']
    df_reg[categoria] = df_reg[categoria].astype('category')
    df_reg[categoria] = df_reg[categoria].cat.reorder_categories(['regular', 'facil','dificil'])
    models = []
    inter = '%s*%s + %s*%s' % (variable1,categoria,variable2,categoria)
    for formula in formulas:
        formula_ = formula % inter
        mod = ols(formula = formula_, data = df_reg).fit(cov_type = 'cluster',
                                                         cov_kwds={'groups': df_reg['equipo']})
        models.append(mod)
    regressors = ['Intercept',
                  variable1,
                  variable2,
                  categoria + '[T.facil]',
                  categoria + '[T.dificil]', 
                  variable1 + ':' + categoria + '[T.facil]',
                  variable1 + ':' + categoria + '[T.dificil]',
                  variable2 + ':' + categoria + '[T.facil]',
                  variable2 + ':' + categoria + '[T.dificil]']
    tble = summary_col(models,
                       regressor_order = regressors,
                       drop_omitted=True,
                       stars = True,
                       float_format = '%0.3f',
                       model_names = ['Base','EF Equipo','EF Torneo', 'EFs Equipo y Torneo'],
                       info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                                  'R2':lambda x: "{:.2f}".format(x.rsquared),
                                  'R2a': lambda x: "{:.2f}".format(x.rsquared_adj),
                                  'LL': lambda x: "{:.2f}".format(x.llf)})
    return tble.tables[0]
```
