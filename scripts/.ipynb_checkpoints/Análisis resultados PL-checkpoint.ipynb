# Análisis resultados premier league

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def HistogramaAleatoriedad(df, dificultad, definicion, columns = ['Inglaterra']):
    cols = ['Indicador'] + [c for c in df.columns.tolist() if c in columns]
    for col in cols[1:]:
        plt.hist(df[col].values, edgecolor='k', alpha=0.65)
        plt.axvline(df.mean()[col], color='r', linestyle='dashed', linewidth=1, label ='Media')
        plt.axvline(df.median()[col], color='g', linestyle='dashed', linewidth=1, label = 'Mediana')
        plt.xlabel('Valor indicador')
        plt.ylabel('Frecuencia')
        plt.title('Histograma Aleatoriedad Frac. Partidos %s según definición %s \n %s' % (dificultad, definicion, col))
        plt.legend()
        plt.show()
```

## Aleatoriedad inicio

```python
resultadosdir = os.path.join(os.path.pardir, 'resultados', 'aleatoriedad-inicio')
```

### Aleatoriedad ELO

```python
# Fracción partidos fáciles
df = pd.read_excel(os.path.join(resultadosdir, 'aleatoriedad-elo.xlsx'), sheet_name='Facil')
HistogramaAleatoriedad(df, 'Fácil', 'ELO')
```

```python
# Fracción partidos difíciles
df = pd.read_excel(os.path.join(resultadosdir, 'aleatoriedad-elo.xlsx'), sheet_name='Dificil')
HistogramaAleatoriedad(df, 'Difícil', 'ELO')
```

### Aleatoriedad General

```python
# Fracción partidos fáciles
df = pd.read_excel(os.path.join(resultadosdir, 'aleatoriedad-general.xlsx'), sheet_name='Facil')
for h in range(1,6):
    df_h = df[df['Indicador'].str.contains("Ant. %s" % h)].reset_index(drop=True)
    HistogramaAleatoriedad(df_h, 'Fácil', 'General con %s torneos anteriores' % h)
```

```python
# Fracción partidos difíciles
df = pd.read_excel(os.path.join(resultadosdir, 'aleatoriedad-general.xlsx'), sheet_name='Dificil')
for h in range(1,6):
    df_h = df[df['Indicador'].str.contains("Ant. %s" % h)].reset_index(drop=True)
    HistogramaAleatoriedad(df_h, 'Difícil', 'General con %s torneos anteriores' % h)
```

```python
# Fracción partidos fáciles
df = pd.read_excel(os.path.join(resultadosdir, 'aleatoriedad-localvisita.xlsx'), sheet_name='Facil')
for h in range(1,6):
    df_h = df[df['Indicador'].str.contains("Ant. %s" % h)].reset_index(drop=True)
    HistogramaAleatoriedad(df_h, 'Fácil', 'LV con %s torneos anteriores' % h)
```
